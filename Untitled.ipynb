## Import libraries 

```python
import pandas as pd
import numpy as np
```

## Prepare the dataset

```python
url = 'https://opendata.muenchen.de/dataset/5e73a82b-7cfb-40cc-9b30-45fe5a3fa24e/resource/40094bd6-f82d-4979-949b-26c8dc00b9a7/download/220511_monatszahlenmonatszahlen2204_verkehrsunfaelle.csv'
df = pd.read_csv(url)[df.JAHR!=2022]
```

```python
df['MONAT_DATETIME'] = pd.to_datetime(df[df.MONAT != 'Summe'].MONAT,format='%Y%m')
df[df.MONAT!='Summe'][df.MONATSZAHL=='Alkoholunfälle'][df.AUSPRAEGUNG=='insgesamt'].sort_values(by='MONAT_DATETIME')[['MONAT_DATETIME','WERT']].plot(x='MONAT_DATETIME',y='WERT',figsize=(22,8))
df[df.MONAT!='Summe'][df.MONATSZAHL=='Alkoholunfälle'][df.AUSPRAEGUNG=='Verletzte und Getötete'].sort_values(by='MONAT_DATETIME')[['MONAT_DATETIME','WERT']].plot(x='MONAT_DATETIME',y='WERT',figsize=(22,8))
```

```python
df[df.MONAT!='Summe'][df.MONATSZAHL=='Alkoholunfälle'][df.AUSPRAEGUNG=='insgesamt'][df.JAHR==2000].sort_values(by='MONAT_DATETIME')[['MONAT_DATETIME','WERT']].plot(x='MONAT_DATETIME',y='WERT',figsize=(22,8))
```
